```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score, roc_curve, auc, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import plot_confusion_matrix

from matplotlib.legend_handler import HandlerLine2D

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
dfEmpleados = pd.read_csv("../data/df_empleados_final.csv")
dfEmpleados.shape
```

Luego de haber cargado los datos, procedemos a filtrar solo aquellos empleados cuyo retiro fue por renuncia propia o que aún se encuentren laborando en la empresa

```python
dfEmpModelo = dfEmpleados[(dfEmpleados['CAUSA_NOMBRE'].isna()) | (dfEmpleados['CAUSA_NOMBRE']=='RENUNCIA') ]
```

Con el fin de implementar un modelo de RandomForest y Gradient Boosting, transformamos los diferentes campos categóricos en formato OneHotEncoding. 

```python
dfDummiesEstadoCivil=pd.get_dummies(dfEmpModelo['ID_ESTADO_CIVIL'])
dfDummiesGenero=pd.get_dummies(dfEmpModelo['ID_GENERO'])
dfDummiesGenero.rename(columns={"F":"FEMENINO", "M":"MASCULINO"}, inplace=True)
dfDummiesTipoCargo=pd.get_dummies(dfEmpModelo['TIPO_CARGO'])
dfDummiesTipoCargo.rename(columns={"NS-NR":"CARGO_NSNR"
                                   ,"ADMINISTRATIVO":"CARGO_ADMINISTRATIVO"
                                  ,"APRENDICES":"CARGO_APRENDICES"
                                  ,"ESTRATEGICO":"CARGO_ESTRATEGICO"
                                  ,"OPERATIVO":"CARGO_OPERATIVO"}, inplace=True)
dfDummiesTipoArea=pd.get_dummies(dfEmpModelo['TIPO_AREA'])
dfDummiesTipoArea.rename(columns={"ADMINISTRATIVO":"AREA_ADMIN"
                                 ,"COMERCIAL":"AREA_COMERCIAL"
                                 ,"LOGISTICA":"AREA_LOGISTICA"
                                 ,"OPERATIVO":"AREA_OPERATIVA"
                                 ,"PRACTICANTES":"AREA_PRACTICANTES"
                                 ,"PROTEICOS":"AREA_PROTEICOS"
                                 ,"SOPORTE":"AREA_SOPORTE"}, inplace=True)
dfDummiesTipoNomina=pd.get_dummies(dfEmpModelo['TIPO_NOMINA'])
dfDummiesTipoNomina.rename(columns={"APRENDIZ":"NOMINA_APRENDIZ"
                                   ,"INTEGRAL":"NOMINA_INTEGRAL"
                                   ,"JUBILADO":"NOMINA_JUBILADO"
                                   ,"NS-NR":"NOMINA_NSNR"
                                   ,"TERMINO_FIJO":"NOMINA_FIJO"
                                   ,"TERMINO_INDEFINIDO":"NOMINA_INDEFINIDO"
                                   ,"TRADICIONAL_NOMINAL":"NOMINA_TRADICIONAL"}, inplace=True)

dfTotalDummies=pd.concat([dfDummiesEstadoCivil
                      ,dfDummiesGenero
                      ,dfDummiesTipoCargo
                      ,dfDummiesTipoArea
                      ,dfDummiesTipoNomina], axis=1)
```

Concatenamos el resultado de la transformación a nuestra base filtrada para implementar el modelo:

```python
dfEmpModelo.drop(columns=['ID_ESTADO_CIVIL','ID_GENERO','TIPO_CARGO','TIPO_AREA','TIPO_NOMINA','FECHA','CODIGO_EMPRESA','UBICACION_ANONIMIZADA','CODIGO_LOCALIDAD','FECHA_NACIMIENTO','FECHA_INGRESO','FECHA_INICIO_CONTRATO','ID_CARGO','NIVEL_CARGO','CATEGORIA_CARGO','COMPENSACION_VARIABLE','PORCENT_COMP_VARIABLE','SALARIO','FECHA_FIN_CONTRATO','CAUSA_NOMBRE','TIPO_RETIRO'], inplace=True)
dfEmpFinal=pd.concat([dfEmpModelo,dfTotalDummies], axis=1)
dfEmpFinal['ID_CENTRO_COSTO'] = dfEmpFinal['ID_CENTRO_COSTO'].apply(str).str[:2]
dfEmpFinal = dfEmpFinal[['EDAD', 'ID_CENTRO_COSTO', 'SUBSIDIO_TRANSPORTE', 'TOTAL_SALARIO', 'INDICADOR_VARIACION_SALARIO', 'VARIACION_SALARIO','SENAL_SINDICALIZADO', 'NUM_DIAS_LABORADOS','NUM_DIAS_CONTRATADOS', 'CASADO', 'DIVORCIADO','NO_DEFINIDO', 'SEPARADO', 'SOLTERO', 'UNION_LIBRE', 'VIUDO','FEMENINO', 'MASCULINO', 'CARGO_ADMINISTRATIVO','CARGO_APRENDICES', 'CARGO_ESTRATEGICO', 'CARGO_NSNR','CARGO_OPERATIVO', 'AREA_ADMIN', 'AREA_COMERCIAL','AREA_LOGISTICA', 'AREA_OPERATIVA', 'AREA_PRACTICANTES','AREA_PROTEICOS', 'AREA_SOPORTE', 'NOMINA_APRENDIZ','NOMINA_INTEGRAL', 'NOMINA_JUBILADO', 'NOMINA_NSNR', 'NOMINA_FIJO','NOMINA_INDEFINIDO', 'NOMINA_TRADICIONAL','RETIRO']]
dfEmpFinal.head()
```

## Partición de datos en Entrenamiento y Prueba

Teniendo el DataFrame listo, procedemos a crear nuestros conjuntos de datos para entrenamiento y para prueba. Utilizamos una relación de 80% para entrenamiento y 20% para prueba, estratificado por nuestra variable dependiente y dejando una semilla para su futura reproducción

```python
varIndep = dfEmpFinal[['EDAD', 'ID_CENTRO_COSTO', 'SUBSIDIO_TRANSPORTE', 'TOTAL_SALARIO', 'INDICADOR_VARIACION_SALARIO', 'VARIACION_SALARIO','SENAL_SINDICALIZADO', 'NUM_DIAS_LABORADOS','NUM_DIAS_CONTRATADOS', 'CASADO', 'DIVORCIADO','NO_DEFINIDO', 'SEPARADO', 'SOLTERO', 'UNION_LIBRE', 'VIUDO','FEMENINO', 'MASCULINO', 'CARGO_ADMINISTRATIVO','CARGO_APRENDICES', 'CARGO_ESTRATEGICO', 'CARGO_NSNR','CARGO_OPERATIVO', 'AREA_ADMIN', 'AREA_COMERCIAL','AREA_LOGISTICA', 'AREA_OPERATIVA', 'AREA_PRACTICANTES','AREA_PROTEICOS', 'AREA_SOPORTE', 'NOMINA_APRENDIZ','NOMINA_INTEGRAL', 'NOMINA_JUBILADO', 'NOMINA_NSNR', 'NOMINA_FIJO','NOMINA_INDEFINIDO', 'NOMINA_TRADICIONAL']].values
varDep = dfEmpFinal[['RETIRO']].values

X_train, X_test, y_train, y_test = train_test_split(varIndep
                                                    , varDep
                                                    , test_size=0.20
                                                    , random_state=20201124
                                                    , stratify=varDep)
```

Graficamos la distribución de los registros en entrenamiento y prueba, evidenciando que tenemos datos de retiro para ambos conjuntos

```python
plt.figure(figsize=(10,6))
plt.hist(y_train, label="Train")
plt.hist(y_test, label="Test")
plt.legend()
```

## MODELADO

### Random Forest

Para el uso de este modelo, definimos una grilla de parámetros que nos permitirá utilizar diferentes combinaciones para evaluar nuestro modelo. Utilizamos una validación cruzada de hasta 5 iteraciones, para tener mayor asertividad. 

```python
param_grid = {
    'bootstrap': [True],
    'max_depth': [70, 80, 90, 100, 150],
    'max_features': [3, 6, 8],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}

model_rf = RandomForestClassifier()
grid_search = GridSearchCV(estimator = model_rf, param_grid = param_grid, 
                          cv = 5, n_jobs = -1, verbose = 2)
grid_search.fit(X_train, y_train)
rfc_pred = grid_search.predict(X_test)
```

Luego de la ejecución de la grilla, identificamos los parámetros que mejor resultado dieron, para así entrenar nuestro modelo

```python
grid_search.best_params_
```

Entrenamos y probamos el modelo con los parámetros establecidos:

```python
ranf = RandomForestClassifier(bootstrap=True,
                              max_depth=100,
                              max_features=8,
                              min_samples_leaf=3,
                              min_samples_split=8,
                              n_estimators=200)
ranf.fit(X_train, y_train.ravel())

rfc_pred = ranf.predict(X_test)
```

Identificamos cuáles son las variables con mayor importancia para el modelo

```python
importances = ranf.feature_importances_
indices = np.argsort(importances)[::-1]
plt.figure()
plt.title("Importancia de las características")
plt.bar(range(X_train.shape[1]), importances[indices],
       color="brown", align="center")
xticks_labels = [dfEmpFinal.columns[1+i] for i in indices]
plt.xticks(range(varIndep.shape[1]), xticks_labels, rotation=90)
plt.xlim([-1, varIndep.shape[1]])
plt.show()
```

De acuerdo al modelo, las variables que mejor explicaron nuestra variable objetivo "RETIRO" fueron:
* Número de Días Contratados
* Estado Civil: CASADO
* Cargo: No sabe, No responde
* Indicador de si al empleado le realizaron un ajuste salarial
* Indicador de si el empleado se encuentra sindicalizado

Realizamos nuestra matriz de Confusión para, de manera visual, identificar qué tan acertado estuvo nuestro modelo:

```python
%config InlineBackend.figure_format = 'svg'

fig = plt.figure(figsize=(15,15))

ax1 = fig.add_subplot(1,3,1)
plot_confusion_matrix(ranf, X_test, y_test, cmap = 'Oranges', ax=ax1).im_.colorbar.remove()
plt.title("Random Forest")
```

Sacamos nuestras métricas de interés:

```python
cm_rf = confusion_matrix(y_test,rfc_pred)
total1=sum(sum(cm_rf))
print('Resultado de Métricas para Modelo Random Forest')
accuracy1=(cm_rf[0,0]+cm_rf[1,1])/total1
print ('Accuracy : ', round(accuracy1,3) )

sensitivity1 = cm_rf[0,0]/(cm_rf[0,0]+cm_rf[0,1])
print('Sensibilidad : ', round(sensitivity1,3) )

specificity1 = cm_rf[1,1]/(cm_rf[1,0]+cm_rf[1,1])
print('Especificidad : ', round(specificity1,3) )
```

Teniendo en cuenta que, nuestro modelo de regresión logística en __R__ nos generó los siguientes valores:

* Accuracy: 0.884
* Sensibilidad: 0.985
* Especificidad: 0.377

Con el resultado del RandomForest, mejoramos el __accuracy__ y la __especificidad__ disminuyendo un poco la __sensibilidad__.

### Gradient Boosting

Para este modelo, hemos creado una lista de estimadores que nos permitirá identificar de manera rápida cuál será el mejor estimador para el modelo

```python
n_estimators = [2, 4, 8, 10, 15, 20, 30, 40, 50, 60, 80, 95, 100, 150, 200]
train_results = []
test_results = []
for estimator in n_estimators:
    model = GradientBoostingClassifier(n_estimators=estimator)
    model.fit(X_train, y_train.ravel())
    train_pred = model.predict(X_train)
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train, train_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    train_results.append(roc_auc)
    y_pred = model.predict(X_test)
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    test_results.append(roc_auc)
plt.figure(figsize = (10, 6))
line1, = plt.plot(n_estimators, train_results, label='AUC en Entrenamiento')
line2, = plt.plot(n_estimators, test_results, label='AUC en Test')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.arrow(95, 0.8, 0, 0.07, head_length=0.02, fc='k', ec='k')
plt.text(78, 0.77, 'Punto de balance')
plt.ylabel('AUC score')
plt.xlabel('Número de Estimadores')
plt.title('Curva ROC - Gradient Boosting')
plt.show()
```

Nuestra curva ROC nos indica que, el número de estimadores con mejor balance es de __95__. Dado esto, entrenamos nuestro modelo con los datos de prueba y los probamos

```python
model_gb = GradientBoostingClassifier(n_estimators=95)
model_gb.fit(X_train, y_train.ravel())
y_pred_gb = model_gb.predict(X_test)
false_positive_rate_gb, true_positive_rate_gb, thresholds = roc_curve(y_test, y_pred_gb)
roc_auc_gb = auc(false_positive_rate_gb, true_positive_rate_gb)
```

Realizamos nuestra matriz de Confusión para, de manera visual, identificar qué tan acertado estuvo nuestro modelo:

```python
%config InlineBackend.figure_format = 'svg'

fig = plt.figure(figsize=(15,15))

ax1 = fig.add_subplot(1,3,1)
plot_confusion_matrix(model_gb, X_test, y_test, cmap = 'Oranges', ax=ax1).im_.colorbar.remove()
plt.title("Gradient Boosting")
```

Sacamos nuestras métricas de interés:

```python
cm_gb = confusion_matrix(y_test,y_pred_gb)
total1=sum(sum(cm_gb))
print('Resultado de Métricas para Modelo Gradient Boosting')
accuracy1=(cm_gb[0,0]+cm_gb[1,1])/total1
print ('Accuracy : ', round(accuracy1,3))

sensitivity1 = cm_gb[0,0]/(cm_gb[0,0]+cm_gb[0,1])
print('Sensibilidad : ', round(sensitivity1,3) )

specificity1 = cm_gb[1,1]/(cm_gb[1,0]+cm_gb[1,1])
print('Especificidad : ', round(specificity1,3))
```

En comparación con nuestro modelo de Random Forest, podemos ver que, se mejoraron las métricas de __Accuracy__ y __Especificidad__ mientras que la __Sensibilidad__ disminuyó un poco, pero sigue estando en valores sobresalientes.

Concluimos que, el modelo con las mejores métricas dados los datos de prueba, fue el de __Gradient Boosting__ con __95__ número de estimadores. 

```python
#Componente competitivo
prediccion_Retiro = model_gb.predict(varIndep)
dfKaggle = dfEmpModelo[['CEDULA_ANONIMIZADA','RETIRO']]
dfKaggle['PREDICCION_RETIRO'] = prediccion_Retiro
dfKaggle.to_csv(r'../data/Entrega_Pronostico_Empleados_Parcial2.csv', index=False)
```
